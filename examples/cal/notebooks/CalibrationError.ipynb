```python
def transfer_function_fractional_amp_error(f):
    alpha = -0.2+0.2*f/1000.0
    return alpha
    
def transfer_function_phase_error(f):
    delta_phic = 0.0 # phase shift in radians
    delta_tc = 10.0 # time shift in seconds
    return delta_phic - 2*np.pi*f*delta_tc

def k_error(f):
    deltaabya = transfer_function_fractional_amp_error(f)
    deltaphi = transfer_function_phase_error(f)
    return (1.0+deltaabya)*np.exp(1.0j*deltaphi)
    #return 1.0+deltaabya


def adjust_frequency_series_with_new_transfer_function(strain_tilde):
    #k_array = k_error(strain_tilde.sample_frequencies)
    k_array = np.array([k_error(f) for f in strain_tilde.sample_frequencies])
    print k_array[0], k_array[-1]
    adjusted = strain_tilde * k_array
    return pycbc.types.FrequencySeries(adjusted, delta_f=strain_tilde.delta_f)


def adjust_strain_with_new_transfer_function(strain):
    
    epoch = strain.start_time
    
    # Fourier transform the time series, then return it as a FrequencySeries.
    # This does not set the epoch of the FrequencySeries. You have to do that manually.
    st = strain.to_frequencyseries()
    print epoch, st.epoch
    
    st_adjusted = adjust_frequency_series_with_new_transfer_function(st)
    s_adjusted = st_adjusted.to_timeseries()
    
    s_adjusted.start_time = epoch
    
    return s_adjusted
```

# Modify the strain data then write it to a .gwf frame file for each alpha

```python
# get list of alphas (the amplitude modulation factor
# alphas is a list of alphas to modulate h(t) by
alpha_min = 0.4
alpha_max = 1.4
num_alpha = 22
alphas = np.linspace(alpha_min, alpha_max, num_alpha)
```

```python
# loop over alpha (the amplitude modulation factor)
for alpha in alphas:

    # model for amplitude modulation
    # this is a simple model that just adjusts the amplitude by a factor alpha
    # there is no frequency-dependence
    amp = alpha

    # apply modulation
    logging.info('Modulate timeseries with alpha set to '+str(alpha)+'...')
    strain_alpha = amp * strain.numpy()

    # create a new Timeseries instance since modulation casts strain_alpha as an array
    # and the write_frame function requires a Timeseries
    strain_alpha = types.TimeSeries(strain_alpha, delta_t=strain.delta_t,
                                          epoch=strain.start_time)

    # Check if output dir exists.
    # If not, create it and set output frame filename.
    tag                 = 'ALPHA_'+str('%.2f'%alpha).replace('.', '_')
    output_dir          = 'frames/'+tag
    output_filename     = output_dir+'/'+ifo+'-'+tag+'-'+str(start_time)+'-'+str(end_time-start_time)+'.gwf'
    # the new channel name modified to state that it has an adjusted value of alpha
    channel_name_alpha = channel_name+'_'+tag
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # save timeseries to frame file
    logging.info('Saving timeseries to frame files...')
    frame.write_frame(output_filename, channel_name_alpha, strain_alpha)
```

```python
def run_pycbc_inspiral(tag, channel_name_alpha, TRIG_START_TIME, DURATION, BUFFER):
    """
    BUFFER: INT
        GPS_START_TIME will be BUFFER seconds befor TRIG_START_TIME.
    """
    # gps start and end times for recording triggers
    # !!!Why does this interval need to be less than the GPS times?!!!
    TRIG_END_TIME = TRIG_START_TIME + DURATION
    # gps start and end times of the data you want to look at
    GPS_START_TIME = TRIG_START_TIME - BUFFER
    GPS_END_TIME = TRIG_END_TIME + BUFFER
    
    ############ Names of files #############
    
    # The frame file to get the GW data from
    FRAME_FILE=glob.glob('frames/'+tag+'/*.gwf')[0]
    #FRAME_FILE='frames/'+tag+'-'+str(GPS_START_TIME)+'-'+str(DURATION)+'.gwf'
    
    # file name for the output of pycbc_inspiral
    OUTPUT_FILE='output/L1-INSPIRAL_PYCBC_'+tag+'-'+str(GPS_START_TIME)+'-'+str(DURATION)+'.xml.gz'
    
    # xml file that contains the template bank
    BANK_FILE='tmpltbank/L1-TMPLTBANK_SNGL-1102660163-1920.xml.gz'
    
    # file name for the PSD calculated by pycbc_inspiral
    PSD_OUTPUT_FILE='output/L1-PSD_PYCBC_'+tag+'-'+str(GPS_START_TIME)+'-'+str(DURATION)+'.txt'
    
    ######## Do matched filtering with pycbc_inpsiral ########
    command = ('pycbc_inspiral --segment-end-pad 64 \
               --segment-length 256 \
               --segment-start-pad 64 \
               --psd-estimation median \
               --psd-segment-length 256 \
               --psd-segment-stride 128 \
               --psd-inverse-length 16 \
               --pad-data 8 \
               --sample-rate 4096 \
               --low-frequency-cutoff 40 \
               --strain-high-pass 30 \
               --filter-inj-only  \
               --processing-scheme cpu \
               --cluster-method window \
               --cluster-window 10 \
               --approximant SPAtmplt \
               --order 7 \
               --snr-threshold 5.5 \
               --channel-name '+channel_name_alpha+' \
               --gps-start-time '+str(GPS_START_TIME)+' \
               --gps-end-time '+str(GPS_END_TIME)+' \
               --trig-start-time '+str(TRIG_START_TIME)+' \
               --trig-end-time '+str(TRIG_END_TIME)+' \
               --output '+OUTPUT_FILE+' \
               --frame-files '+FRAME_FILE+' \
               --bank-file '+BANK_FILE+' \
               --psd-output '+PSD_OUTPUT_FILE+' \
               --user-tag PYCBC \
               --zpk-z 100. 100. 100. 100. 100. \
               --zpk-p 1. 1. 1. 1. 1. \
               --zpk-k 1e-10 \
               --normalize-strain 3995. \
               --verbose')
    
    proc = subprocess.Popen(command, shell=True)
    proc.wait()
```

```python
# Create content handler (?)

class DefaultContentHandler(glue.ligolw.lsctables.ligolw.LIGOLWContentHandler):
    pass
glue.ligolw.lsctables.use_in(DefaultContentHandler)
```

```python
# initializations
inj_times = [1102660188, 1102661935] #1102660292, 1102660751, 1102660794]

# get all trigger files
filenames = glob.glob('output/*INSPIRAL*.xml.gz')
```

```python
def get_alpha_from_file_name(filename):
    alpha = float('.'.join(filename.split('-')[1].split('_')[-2:]))
    return alpha

def get_trigger_corresponding_to_injection(inspiral_xml, inj_time, dt):
    
    # Get a list of sngl_inspiral trigger events from the inspiral_xml file object
    sngl_table = glue.ligolw.table.get_table(inspiral_xml, glue.ligolw.lsctables.SnglInspiralTable.tableName)
    
    for row in sngl_table:
        # row.end_time is GPS time of coalesence in seconds
        # row.end_time_ns is the decimal in ns
        tc = row.end_time+1.0e-9*row.end_time_ns
        
        # If you find the injection, exit the for loop
        if tc>inj_time-dt and tc<inj_time+dt:
            return tc, row
    
    # You made it out of the for loop, so no trigger corresponds to the time of the injected event
    raise Exception, 'Injection corresponding to inj_time ='+str(inj_time)+' +/- '+str(dt)+' not found.'
```

```python
inj_time = inj_times[0]
dt = 5.0

alpha_params = []
for filename in filenames:
    
    # Get alpha
    alpha = get_alpha_from_file_name(filename)
        
    # Load xml document containing properties for all the triggers
    inspiral_xml = utils.load_filename(filename, contenthandler=DefaultContentHandler)
    # Get correct injection
    tc, recovered_injection = get_trigger_corresponding_to_injection(inspiral_xml, inj_time, dt=dt)
    
    alpha_params.append([alpha, recovered_injection.snr, recovered_injection.eff_distance])
    
# Sort the list by alpha value
#print alpha_params
alpha_params.sort(key=lambda x: x[0])
#print alpha_params
alpha_params_array=np.array(alpha_params)
```
