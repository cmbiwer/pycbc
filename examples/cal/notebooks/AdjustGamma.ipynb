# Run pycbc for data varied as a function of gamma

```python
#Epoch 2
i=1

freqarray = Cepochlist[i][:, 0]
Carray = Cepochlist[i][:, 1]
Darray = Depochlist[i][:, 1]
Aarray = Aepochlist[i][:, 1]
```

```python
ifo          = 'L1'
channel_name = 'L1:OAF-CAL_DARM_DQ'
BANK_FILE='tmpltbank/L1-TMPLTBANK_SNGL-1102660163-1920.xml.gz'

TRIG_START_TIME = 1102660163
DURATION = 1920
BUFFER = 64


# get list of alphas (the amplitude modulation factor
# alphas is a list of alphas to modulate h(t) by
gamma_min = 0.8
gamma_max = 1.2
Nalpha = 21
gamma_array = np.linspace(gamma_min, gamma_max, Nalpha)


# Loop over gamma values
for gamma in gamma_array:
    
    strain_gamma = transfer.adjust_strain_with_new_transfer_function(strain, freqarray, Carray, Darray, Aarray, gamma=gamma)
    
    # Check if output dir exists.
    # If not, create it and set output frame filename.
    tag                 = 'REGAMMA_'+str('%.3f'%gamma).replace('.', '_')
    frame_dir          = 'frames/REGAMMA'
    frame_filename     = frame_dir+'/'+ifo+'-'+tag+'-'+str(start_time)+'-'+str(end_time-start_time)+'.gwf'
    
    # Create the frame_dir if it doesn't already exist
    if not os.path.exists(frame_dir):
        os.makedirs(frame_dir)
    
    # Write the frame file
    frame.write_frame(frame_filename, channel_name, strain_gamma)
    
    ########## Run pycbc_inspiral on the frame file ###########
    
    # gps start and end times of the data you want to look at
    GPS_START_TIME = TRIG_START_TIME - BUFFER
    GPS_END_TIME = TRIG_START_TIME + DURATION + BUFFER
    
    FRAME_FILE = frame_filename
    
    output_dir = 'output/REGAMMA'
    OUTPUT_FILE = output_dir+'/L1-INSPIRAL_PYCBC_'+tag+'-'+str(GPS_START_TIME)+'-'+str(DURATION)+'.xml.gz'
    PSD_OUTPUT_FILE=output_dir+'/L1-PSD_PYCBC_'+tag+'-'+str(GPS_START_TIME)+'-'+str(DURATION)+'.txt'
    
    # Create the output_dir if it doesn't already exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    triggers.run_pycbc_inspiral(FRAME_FILE, channel_name, BANK_FILE, OUTPUT_FILE, PSD_OUTPUT_FILE, TRIG_START_TIME, DURATION, BUFFER)
```

```python
def get_gamma_from_file_name(filename):
    # Get the integer part and decimal part then combine into a single string
    i, d = filename.split('_')[-2:]
    gamma = i + '.' + d.split('-')[0]
    return float(gamma)
```

```python
# There were 2 hardware injections with these end times:
inj_times = [1102660188, 1102661935]
# get all trigger files
filenames = glob.glob('output/REGAMMA/*INSPIRAL*.xml.gz')

inj_time = inj_times[0]
dt = 5.0

gamma_params = []
for filename in filenames:
    
    # Get gamma
    gamma = get_gamma_from_file_name(filename)
        
    trig_time, trig = triggers.get_trigger_corresponding_to_injection(filename, inj_time, dt)
    
    gamma_params.append([gamma, trig.snr, trig.eff_distance])
    
# Sort the list by gamma value
gamma_params.sort(key=lambda x: x[0])
regamma_params_array=np.array(gamma_params)
```

```python
fig = plt.figure(figsize=(8, 6))
axes = fig.add_subplot(111)
axes.plot(regamma_params_array[:, 0]-1, regamma_params_array[:, 1], ls='-', lw=2, c='b', label='T='+str(inj_time))
axes.set_xlim([-0.2, 0.2])
axes.set_ylim([0.0, 40.0])
axes.set_xlabel(r'$\Re(\gamma)-1$', fontsize=16)
axes.set_ylabel('SNR', fontsize=16)
axes.set_xticklabels(axes.get_xticks(), fontsize=14)
axes.set_yticklabels(axes.get_yticks(), fontsize=14)
axes.minorticks_on()
axes.tick_params(which='major', width=2, length=8)
axes.tick_params(which='minor', width=2, length=4)
axes.legend(loc='lower right')

fig = plt.figure(figsize=(8, 6))
axes = fig.add_subplot(111)
axes.plot(regamma_params_array[:, 0]-1, regamma_params_array[:, 2], ls='-', lw=2, c='b', label='T='+str(inj_time))
axes.set_xlim([-0.2, 0.2])
axes.set_ylim([0.0, 150])
axes.set_xlabel(r'$\Re(\gamma)-1$', fontsize=16)
axes.set_ylabel(r'$D_{\rm eff}$ (Mpc)', fontsize=16)
axes.set_xticklabels(axes.get_xticks(), fontsize=14)
axes.set_yticklabels(axes.get_yticks(), fontsize=14)
axes.minorticks_on()
axes.tick_params(which='major', width=2, length=8)
axes.tick_params(which='minor', width=2, length=4)
axes.legend(loc='lower right')
```
